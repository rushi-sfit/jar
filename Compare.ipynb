```python
import pandas as pd
import matplotlib.pyplot as plt
apple_df=pd.read_csv('D:/SMA EXP/sentiment_analysis_apple.csv')
samsung_df=pd.read_csv('D:/SMA EXP/sentiment_analysis_samsung.csv')
apple_df.rename(columns={'sentiment_label':'Sentiment'},inplace=True)
samsung_df.rename(columns={'sentiment_label':'Sentiment'},inplace=True)
```

```python
plt.figure(figsize=(8,6))
apple_df['Sentiment'].value_counts().plot(kind='bar',color=['green','red','gray'],edgecolor='black')
plt.title('Apple reviews')
plt.xlabel('Frequency')
plt.ylabel('Sentiment')
plt.show()
```

```python
plt.figure(figsize=(8,6))
samsung_df['Sentiment'].value_counts().plot(kind='bar',color=['green','red','gray'],edgecolor='black')
plt.title('Samsung reviews')
plt.xlabel('Frequency')
plt.ylabel('Sentiment')
plt.show()
```

```python
#Comparative
def calculate_sentiment_count(df):
    sentiment_counts=df['Sentiment'].value_counts()
    counts={
        'Positive':sentiment_counts['positive'] if 'positive' in  sentiment_counts else 0,
        'Negative':sentiment_counts['negative'] if 'negative' in  sentiment_counts else 0,
        'Neutral':sentiment_counts['neutral'] if 'neutral' in  sentiment_counts else 0
            }
    return counts

apple_sentiment_count=calculate_sentiment_count(apple_df)
samsung_sentiment_count=calculate_sentiment_count(samsung_df)

combined_df=pd.DataFrame({
    'Apple':apple_sentiment_count,
    'Samsung':samsung_sentiment_count
    })
```

```python
plt.figure(figsize=(8,6))
combined_df.plot(kind='bar',color=['skyblue','gray'],edgecolor='black')
plt.title('Samsung reviews')
plt.xlabel('Sentiment')
plt.ylabel('Frequency')
plt.show()
```
