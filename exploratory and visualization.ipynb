```python
 from matplotlib import pyplot as plt
 import seaborn as sns
 import numpy as num
 import pandas as pd
 data = pd.read_csv('D:/SMA EXP/37_exp6_data.csv')
 data.head()
```

```python
print("Mean: ",data.mean()['numeric_id_2'])
print("Median: ",data.median()['numeric_id_2'])
print("Mode: ",data.mode())
```

```python
print("Variance: ",data.var()['numeric_id_2'])
print("Standard Deviation: ",data.std()['numeric_id_2'])
print("Skewness: ",data.skew()['numeric_id_2'])
print("Kurtosis: ",data.kurt()['numeric_id_2'])
```

```python
sns.boxplot(x = 'numeric_id_2', data = data)
plt.title('Boxplot of numeric_id_2')
plt.show()
```

```python
plt.hist(x = 'numeric_id_2', data = data)
plt.title('Bar Graph of numeric_id_2')
plt.show()
```

```python
plt.figure(figsize=(17,6))
sns.scatterplot(x=data["numeric_id_1"], y=data["numeric_id_2"], data=data)
plt.show()
```

```python
plt.figure(figsize=(17,6))
sns.lineplot(x=data["numeric_id_2"], y=data["numeric_id_1"], data=data)
plt.show()
```

```python
sns.countplot(x='numeric_id_2', hue='numeric_id_1', data=data)
plt.show()
```
